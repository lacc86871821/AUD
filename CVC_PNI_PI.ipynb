```python
import pandas as pd
df = list(uploaded.keys())[0]
df_boxplot = pd.read_excel(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets, Layout
from IPython.display import display

# Widget para ajustar el rango del eje y
y_range = widgets.FloatRangeSlider(
    value=[-20000, 20000],
    min=-20000,
    max=20000,
    step=1000,
    description='Rango del eje y:',
    layout=Layout(width='50%')
)

# Widget de lista de chequeo para seleccionar los datos a mostrar
materiales = df_boxplot["Texto breve de material"].unique()
material_selection = widgets.SelectMultiple(
    options=materiales,
    value=tuple(materiales),
    description='Materiales:',
    layout=Layout(width='50%')
)

# Widget de lista de selecciÃ³n para el tipo de pÃ©rdida
tipos_perdida = df_boxplot["Tipo de Perdida"].unique()
tipo_perdida_selection = widgets.SelectMultiple(
    options=tipos_perdida,
    value=tuple(tipos_perdida),
    description='Tipos de PÃ©rdida:',
    layout=Layout(width='50%')
)

# Widget de lista de selecciÃ³n para el mes
meses = df_boxplot["Mes"].unique()
mes_selection = widgets.SelectMultiple(
    options=meses,
    value=tuple(meses),
    description='Meses:',
    layout=Layout(width='50%')
)

def plot_boxplot(y_range, selected_materiales, selected_tipos_perdida, selected_meses):
    df_filtered = df_boxplot[df_boxplot["Texto breve de material"].isin(selected_materiales) &
                            df_boxplot["Tipo de Perdida"].isin(selected_tipos_perdida) &
                            df_boxplot["Mes"].isin(selected_meses)]

    if not df_filtered.empty:
        # Crea una paleta de colores personalizada para los materiales seleccionados
        color_palette = sns.color_palette("Set3", len(selected_materiales))

        sns.set(style="whitegrid")
        plt.figure(figsize=(12, 8))
        ax = sns.boxplot(data=df_filtered, x="Texto breve de material", y="Cantidad", palette=color_palette)
        plt.xticks(rotation=90)
        plt.title("GrÃ¡fico de Caja y Bigotes de Cantidad por Material, Tipo de PÃ©rdida y Mes")

        # Ajusta el rango del eje y segÃºn la selecciÃ³n del usuario
        plt.ylim(y_range[0], y_range[1])

        plt.show()
    else:
        print("No se encontraron datos para los materiales, tipos de pÃ©rdida y meses seleccionados.")

# Widget interactivo que llama a la funciÃ³n plot_boxplot
interact(plot_boxplot, y_range=y_range, selected_materiales=material_selection, selected_tipos_perdida=tipo_perdida_selection, selected_meses=mes_selection)

# Muestra los widgets
display(y_range, material_selection, tipo_perdida_selection, mes_selection)
```
